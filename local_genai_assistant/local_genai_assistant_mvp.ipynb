```python
# ✅ Step 1: Install dependencies
!pip install -q crewai llama-cpp-python PyMuPDF

# ✅ Step 2: Imports
import fitz  # PyMuPDF
import textwrap
from google.colab import files, drive
from llama_cpp import Llama
from crewai import Agent, Task, Crew
from crewai.llms.base_llm import BaseLLM

# ✅ Step 3: Upload PDF
uploaded = files.upload()
pdf_filename = next(iter(uploaded))

# ✅ Step 4: Extract PDF Text
def extract_pdf_text(pdf_path):
    try:
        doc = fitz.open(pdf_path)
        text = ""
        for page_num, page in enumerate(doc, 1):
            page_text = page.get_text()
            text += f"\n--- Page {page_num} ---\n{page_text}"
        return text.strip()
    except Exception as e:
        print(f"❌ Error reading PDF: {e}")
        return ""

pdf_text = extract_pdf_text(pdf_filename)

# ✅ Step 5: Mount Google Drive
drive.mount('/content/drive')

# ✅ Step 6: Load Local Mistral .gguf Model
MODEL_PATH = "/content/drive/MyDrive/mistral-7b-instruct.Q4_K_M.gguf"  # 👈 Update if needed

llama_model = Llama(
    model_path=MODEL_PATH,
    n_ctx=4096,
    n_threads=8,
    use_mlock=False
)

# ✅ Step 7: Custom wrapper for Mistral chat format
class MyLocalLLM(BaseLLM):
    def __init__(self, llama_model):
        self._llama = llama_model
        self.max_tokens = 512

    def call(self, messages: list, **kwargs) -> str:
        # Expecting [{"role": "user", "content": "..."}]
        prompt = self.convert_to_mistral_prompt(messages)
        result = self._llama(prompt=prompt, max_tokens=self.max_tokens)
        return result['choices'][0]['text'].strip()

    def convert_to_mistral_prompt(self, messages: list) -> str:
        prompt = ""
        for msg in messages:
            if msg["role"] == "user":
                prompt += f"[INST] {msg['content']} [/INST]"
            elif msg["role"] == "assistant":
                prompt += msg["content"]
        return prompt

    def supports_stop_words(self) -> bool:
        return False

# ✅ Step 8: Initialize LLM
llm = MyLocalLLM(llama_model)

# ✅ Step 9: Create Agent
reader_agent = Agent(
    role="Document Analyst",
    goal="Extract actionable items and deadlines from business documents",
    backstory="An expert at reading documents and summarizing actionable items for teams.",
    llm=llm
)

# ✅ Step 10: Chunk PDF
def chunk_text(text, max_chars=3000):
    return textwrap.wrap(text, max_chars, break_long_words=False, break_on_hyphens=False)

chunks = chunk_text(pdf_text)

# ✅ Step 11: Create Tasks
tasks = []
for i, chunk in enumerate(chunks):
    tasks.append(Task(
        description=f"""
Analyze the following document section and extract:

• Actionable tasks
• Responsibilities
• Deadlines (if mentioned)
• Follow-up actions

--- Start of Chunk {i+1} ---
{chunk}
--- End of Chunk {i+1} ---
""",
        expected_output="A bullet-point list of all tasks, responsibilities, deadlines, and follow-up actions.",
        agent=reader_agent
    ))

# ✅ Step 12: Run CrewAI
crew = Crew(agents=[reader_agent], tasks=tasks)
final_output = crew.kickoff()

# ✅ Step 13: Display Results
print("\n✅ FINAL EXTRACTED TASKS:\n")
if isinstance(final_output, list):
    for i, result in enumerate(final_output):
        print(f"--- 📄 Chunk {i+1} Results ---\n{result}\n")
else:
    print(final_output)
```

```python
!pip install -q transformers timm

from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import requests

# Load model and processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# ✅ Working image URL (dog on snow)
url = "https://images.pexels.com/photos/1108099/pexels-photo-1108099.jpeg"

# Download and open image
response = requests.get(url, stream=True)
image = Image.open(response.raw).convert('RGB')

# Run captioning
inputs = processor(image, return_tensors="pt")
output = model.generate(**inputs)
print(processor.decode(output[0], skip_special_tokens=True))
```

```python
import json

def save_memory(key, value):
    memory = json.load(open("memory.json")) if os.path.exists("memory.json") else {}
    memory[key] = value
    json.dump(memory, open("memory.json", "w"))

def load_memory(key):
    memory = json.load(open("memory.json"))
    return memory.get(key, "No memory found")
```
