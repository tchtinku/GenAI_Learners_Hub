```python
import streamlit as st
from transformers import pipeline

st.title("🚀 LinkedIn Post Generator")
achievement = st.text_input("Your achievement (e.g., 'Led a team of 5'):")
tone = st.selectbox("Tone:", ["Professional", "Casual", "Motivational"])

if st.button("Generate Post"):
    generator = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.1")
    prompt = f"Generate a {tone} LinkedIn post about {achievement}. Include hashtags."
    post = generator(prompt, max_length=300)[0]['generated_text']
    st.write(post)
```
